```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dados.csv')
```

```python
df
```

```python
# Informações sobre a média(mean), mínimo(min) e máximo(max)
df.describe()
```

```python
# Agrupar os dados de raça/etnia e math score
df2 = df.groupby('race/ethnicity')[['math score']].mean()
df2
```

```python
# Gráfico de raça/etnia e math score
df2.plot(kind= 'line')
```

```python
# Agrupar os dados escolaridade e writing score
df3 = df.groupby('parental level of education')[['writing score']].mean()
df3
```

```python
# Gráfico de escolaridade e writing score
df3.plot(kind= 'bar')
```

```python
# Agrupar o ngc de curso preparatório e reading score
df4 = df.groupby('test preparation course')[['reading score']].mean()
df4
```

```python
# Gráfico curso preparatório e reading score
df4.plot(kind= 'barh')
```
